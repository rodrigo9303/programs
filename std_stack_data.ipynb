# The purpose of this notebook is to make the stacking from BOSS data and compare the stacking between Hélion's and Rodrigo's, including the standar desviation for the stacking of Rodrigo's.

We first import the packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import fitsio 
%matplotlib inline
```

Then we upload the files with the data of the stacking for mocks and BOSS DR9 data of Rodrigo's and the data of Hélion's for mocks and BOSS DR14 data. Note that Rodrigo's data is for a redshift bin from 2.6 to 2.8, while Helion's data is for a redshift from 2.0 to 4.288 (I'm not sure about this).

```python
wavemr,fluxmr=np.loadtxt('mock_2.6_z_2.8.dat',skiprows=2,unpack=True) # data of mock of Rodrigo's
wavedr,fluxdr=np.loadtxt('boss_2.6<z<2.8.dat',skiprows=2,unpack=True) # data of boss data of Rodrigo's
wavemh,fluxmh=np.loadtxt('Mock.dat',skiprows=2,unpack=True)           # data of mock of Hélion's
wavedh,fluxdh=np.loadtxt('Data.dat',skiprows=2,unpack=True)           # data of boss data of Hélion's 
stdmockr = np.loadtxt('std_mock_2.6_z_2.8.dat',skiprows=1,unpack=True) # data of standar desviation of mock of Rodrigo's
stddatar = np.loadtxt('std_boss_2.6<z<2.8.dat',skiprows=1,unpack=True) # data of standar desviation of boss Rodrigo's
```

Then we plot of data for mocks.

```python
plt.figure(figsize=(15,9))
plt.title('Stacking',fontsize = 18)
plt.plot(wavemr,312.5*fluxmr,'o-', color='r',label='Mocks-Rodrigo',linewidth=2.0,alpha=0.5)
#plt.plot(wavedr,fluxdr,'o-', color='b',label='Mocks-Rodrigo',linewidth=2.0,alpha=1)
plt.plot(wavemh,fluxmh,'o-', color='b',label='Mocks-Helion',linewidth=2.0,alpha=0.5)
plt.xlim(1039,1200)
plt.ylim(0.7,1.8)
plt.xlabel('Wavelength$_{R.F.}$  ($\AA$)',fontsize = 18)
plt.ylabel(r'$\mathrm{\overline{Normalized \enspace Flux}}$',fontsize = 18)#     ($10^{-17} ergs/s/cm^{2}/\AA$)',fontsize = 18)
plt.legend(fontsize='xx-large')
plt.grid()
plt.savefig('mocks_RH.png')
```

```python
plt.figure(figsize=(15,9))
plt.title('Stacking',fontsize = 18)
plt.plot(wavemr,312.5*fluxmr,'o-', color='r',label='Mocks-Rodrigo',linewidth=2.0,alpha=0.5)
plt.fill_between(wavemr,312.5*fluxmr+312.5*stdmockr[:712], 312.5*fluxmr-312.5*stdmockr[:712], label='std', color = 'y', alpha=0.5)
#plt.plot(wavedr,fluxdr,'o-', color='b',label='Mocks-Rodrigo',linewidth=2.0,alpha=1)
plt.plot(wavemh,fluxmh,'o-', color='b',label='Mocks-Helion',linewidth=2.0,alpha=0.5)
plt.xlim(1039,1200)
plt.ylim(0.7,1.8)
plt.xlabel('Wavelength$_{R.F.}$  ($\AA$)',fontsize = 18)
plt.ylabel(r'$\mathrm{\overline{Normalized \enspace Flux}}$',fontsize = 18)#     ($10^{-17} ergs/s/cm^{2}/\AA$)',fontsize = 18)
plt.legend(fontsize='xx-large')
plt.grid()
plt.savefig('mocks_RH_std.png')
```

```python
plt.figure(figsize=(15,9))
plt.title('Stacking',fontsize = 18)
plt.plot(wavedr,278*fluxdr,'o-', color='r',label='Boss-Rodrigo',linewidth=2.0,alpha=0.5)
plt.fill_between(wavedr,278*fluxdr+278*stddatar[:666], 278*fluxdr-278*stddatar[:666], label='std', color = 'y', alpha=0.5)
#plt.plot(wavedr,fluxdr,'o-', color='b',label='Mocks-Rodrigo',linewidth=2.0,alpha=1)
plt.plot(wavedh,fluxdh,'o-', color='b',label='Boss-Helion',linewidth=2.0,alpha=0.5)
plt.xlim(1039,1200)
plt.ylim(0.7,1.8)
plt.xlabel('Wavelength$_{R.F.}$  ($\AA$)',fontsize = 18)
plt.ylabel(r'$\mathrm{\overline{Normalized \enspace Flux}}$',fontsize = 18)#     ($10^{-17} ergs/s/cm^{2}/\AA$)',fontsize = 18)
plt.legend(fontsize='xx-large')
plt.grid()
plt.savefig('mocks_RH_std.png')
```

Now I show the standar desviation for the whole spectrum. Note that in the range from ly-$\alpha$ peak under the CIV peak, the stndar desviation is very small, it's due to the fact that I normalize the moscks and BOSS data in this range.

```python
plt.figure(figsize=(15,9))
plt.title('Stacking',fontsize = 18)
plt.plot(wavedr,fluxdr,'o-', color='r',label='Boss-Rodrigo',linewidth=2.0,alpha=0.5)
plt.fill_between(wavedr,fluxdr+stddatar[:666], fluxdr-stddatar[:666], label='std', color = 'y', alpha=0.5)
plt.xlim(1039,1600)
#plt.ylim(0.7,1.8)
plt.xlabel('Wavelength$_{R.F.}$  ($\AA$)',fontsize = 18)
plt.ylabel(r'$\mathrm{\overline{Normalized \enspace Flux}}$',fontsize = 18)#     ($10^{-17} ergs/s/cm^{2}/\AA$)',fontsize = 18)
plt.legend(fontsize='xx-large')
plt.grid()
```
