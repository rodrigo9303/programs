```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1 = np.linspace(0,400,2366) # puntos entre 0 y 100 
x2 = np.linspace(12,415,2366)
x3 = np.linspace(34,465,2366)
x4 = np.linspace(4,408,2366)
y1 = x1**2
y2 = (x2**2) + 5000
y3 = (x3**2) + 14000
y4 = (x4**2) + 22000
nbins = np.linspace(0.,100., 101)#creo los bins
```

```python
x = []
x.append(x1)
x.append(x2)
x.append(x3)
x.append(x4)
xs = np.vstack(x)
```

```python
y = []
y.append(y1)
y.append(y2)
y.append(y3)
y.append(y4)
ys = np.vstack(y)
```

```python
np.min(ys)
```

```python
plt.figure()
for m in range(len(xs)):
    plt.plot(xs[m],ys[m])
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$y = x^2$',fontsize=15)
```

```python
cuent=np.zeros(466)
prom=np.zeros(466)
flu=np.zeros(466)
for i in range(len(xs)):
    for j in range(len(xs[0])):
        cu = int(xs[i][j])
        cuent[cu]+=1
        flu[cu]+=ys[i][j]
        prom = flu/cuent
```
