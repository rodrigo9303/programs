Se puede ver que los datos no forman una distribución gaussiana, más sin embargo, él autor hace el estudio considerando que los datos son gaussianos y hace un análisis de Maximum Likelihood Estimate con esa distribución que pienso no es del todo correcto, ya que la distribución  no es gaussiana.

Con la intención de ver que curva (gaussiana) ajusta mejor a los datos se crearán algunas distribuciones alrededor de los datos.

```python
# Define function that generates values of a normal pdf
def trunc_norm_pdf(xvals, mu, sigma, cutoff):
    '''
    --------------------------------------------------------------------
    Generate pdf values from the normal pdf with mean mu and standard
    deviation sigma. If the cutoff is given, then the PDF values are
    inflated upward to reflect the zero probability on values above the
    cutoff. If there is no cutoff given, this function does the same
    thing as sp.stats.norm.pdf(x, loc=mu, scale=sigma).
    --------------------------------------------------------------------
    INPUTS:
    xvals  = (N,) vector, values of the normally distributed random
             variable
    mu     = scalar, mean of the normally distributed random variable
    sigma  = scalar > 0, standard deviation of the normally distributed
             random variable
    cutoff = scalar or string, ='None' if no cutoff is given, otherwise
             is scalar upper bound value of distribution. Values above
             this value have zero probability
    
    OTHER FUNCTIONS AND FILES CALLED BY THIS FUNCTION: None
    
    OBJECTS CREATED WITHIN FUNCTION:
    prob_notcut = scalar 
    pdf_vals = (N,) vector, normal PDF values for mu and sigma
               corresponding to xvals data
    
    FILES CREATED BY THIS FUNCTION: None
    
    RETURNS: pdf_vals
    --------------------------------------------------------------------
    '''
    if cutoff == 'None':
        prob_notcut = 1.0
    else:
        prob_notcut = sts.norm.cdf(cutoff, loc=mu, scale=sigma)
            
    pdf_vals    = ((1/(sigma * np.sqrt(2 * np.pi)) *
                    np.exp( - (xvals - mu)**2 / (2 * sigma**2))) /
                    prob_notcut)
    
    return pdf_vals

# Plot two different truncated normal distributions
plt.figure(figsize=(8,4))
dist_pts = np.linspace(0, 450, 500)
mu_1 = 300
sig_1 = 30
plt.plot(dist_pts, trunc_norm_pdf(dist_pts, mu_1, sig_1, 450),
         linewidth=2, color='r', label='1: $\mu$=300,$\sigma$=30')
plt.xlabel(r'Total points',fontsize=20)
plt.ylabel(r'Percent of scores',fontsize=20)
plt.legend(loc='upper left')


mu_2 = 400
sig_2 = 60
plt.plot(dist_pts, trunc_norm_pdf(dist_pts, mu_2, sig_2, 450),
         linewidth=2, color='g', label='2: $\mu$=400,$\sigma$=60')
plt.xlabel(r'Total points',fontsize=20)
plt.ylabel(r'Percent of scores',fontsize=20)
plt.legend(loc='upper left')
```

La pregunta es ¿Cuál de todas las distribuciones (gausianas, con parámetros $\mu$ y $\sigma$) es la que maximiza la función $log\enspace likelihood\enspace function$?

Así que ahora se precederá a calcular la $log\enspace likelihood\enspace function$ para estos datos.

```python
# Define log likelihood function for the normal distribution
def log_lik_norm(xvals, mu, sigma, cutoff):
    '''
    --------------------------------------------------------------------
    Compute the log likelihood function for data xvals given normal
    distribution parameters mu and sigma.
    --------------------------------------------------------------------
    INPUTS:
    xvals  = (N,) vector, values of the normally distributed random
             variable
    mu     = scalar, mean of the normally distributed random variable
    sigma  = scalar > 0, standard deviation of the normally distributed
             random variable
    cutoff = scalar or string, ='None' if no cutoff is given, otherwise
             is scalar upper bound value of distribution. Values above
             this value have zero probability
    
    OTHER FUNCTIONS AND FILES CALLED BY THIS FUNCTION:
        norm_pdf()
    
    OBJECTS CREATED WITHIN FUNCTION:
    pdf_vals    = (N,) vector, normal PDF values for mu and sigma
                  corresponding to xvals data
    ln_pdf_vals = (N,) vector, natural logarithm of normal PDF values
                  for mu and sigma corresponding to xvals data
    log_lik_val = scalar, value of the log likelihood function
    
    FILES CREATED BY THIS FUNCTION: None
    
    RETURNS: log_lik_val
    --------------------------------------------------------------------
    '''
    pdf_vals = trunc_norm_pdf(xvals, mu, sigma, cutoff)
    ln_pdf_vals = np.log(pdf_vals)
    log_lik_val = ln_pdf_vals.sum()
    
    return log_lik_val

print('Log-likelihood 1: ', log_lik_norm(pts, mu_1, sig_1, 450))
print('Log-likelihood 2: ', log_lik_norm(pts, mu_2, sig_2, 450))
```

Que son los $log\enspace likelihood$ para las curvas roja (1) y verde (2).

La razón por la cual los $log\enspace likelihood$ son negativos es debido a que los los valores de las curvas roja y verde son números menores a 1 y el $log$ de un número menor a uno es negativo. Además dado que el $likelihood$ es el producto de las probabilidades individuales de cada elemento de los datos (dado que se consideran eventos mutuamente excluyentes), al obtener el $log$ se obtine una suma. Es decir,
\begin{equation}
L(x_{1},x_{2},...,x_{n}|\theta) = \prod_{i=1}^{n}f(x_{i}|\theta).
\end{equation}
Entonces,
\begin{equation}
Ln[L(x_{1},x_{2},...,x_{n}|\theta)] = \sum_{i=1}^{n}Ln[f(x_{i}|\theta)].
\end{equation}

Así que los parámetros obtimos son aquellos que maximizen la probabilidad, en otras palabras, las que maximizen la $log\enspace likelihood\enspace function$. 
\begin{equation}
(\hat{\mu},\hat{\sigma}) = (\mu,\sigma) \hspace{1.5cm} max_{\mu, \sigma} \hspace{0.5cm} Ln[L] = \sum_{i=1}^{n}Ln[f(x_{i}|\theta)]
\end{equation}

La siguiente función calcua el negativo de $Ln(L)$

```python
def crit(params, *args):
    '''
    --------------------------------------------------------------------
    This function computes the negative of the log likelihood function
    given parameters and data. This is the minimization problem version
    of the maximum likelihood optimization problem
    --------------------------------------------------------------------
    INPUTS:
    params = (2,) vector, ([mu, sigma])
    mu     = scalar, mean of the normally distributed random variable
    sigma  = scalar > 0, standard deviation of the normally distributed
             random variable
    args   = length 2 tuple, (xvals, cutoff)
    xvals  = (N,) vector, values of the normally distributed random
             variable
    cutoff = scalar or string, ='None' if no cutoff is given, otherwise
             is scalar upper bound value of distribution. Values above
             this value have zero probability
    
    OTHER FUNCTIONS AND FILES CALLED BY THIS FUNCTION:
        log_lik_norm()
    
    OBJECTS CREATED WITHIN FUNCTION:
    log_lik_val = scalar, value of the log likelihood function
    neg_log_lik_val = scalar, negative of log_lik_val
    
    FILES CREATED BY THIS FUNCTION: None
    
    RETURNS: neg_log_lik_val
    --------------------------------------------------------------------
    '''
    mu, sigma = params
    xvals, cutoff = args
    log_lik_val = log_lik_norm(xvals, mu, sigma, cutoff)
    neg_log_lik_val = -log_lik_val
    
    return neg_log_lik_val
```

Implementando el problema de minimización, se tiene. Nótese que se ha elegido la misma media que ara la curva verde, pero con desviación estándar diferente.

```python
mu_init = 400  # mu_2
sig_init = 70  # sig_2
params_init = np.array([mu_init, sig_init])
mle_args = (pts, 450.0)
results_uncstr = opt.minimize(crit, params_init, args=(mle_args))
mu_MLE, sig_MLE = results_uncstr.x
print('mu_MLE=', mu_MLE, ' sig_MLE=', sig_MLE)
```

Que son justo los parámetros que maximizan el $likelihood$, para nuetros datos.

Veamos la información que ha calculado $minimize$

```python
results_uncstr
```

### Grafiquemos el resultado y veamos cómo se compara con las curvas anteriores.

```python
# Plot the histogram of the data
plt.figure(figsize=(9,5))
count, bins, ignored = plt.hist(pts, num_bins, density=True,
                                edgecolor='k')
plt.title('Econ 381 scores: 2011-2012', fontsize=20)
plt.xlabel(r'Total points',fontsize=20)
plt.ylabel(r'Percent of scores',fontsize=20)
plt.xlim([0, 550])  # This gives the xmin and xmax to be plotted"

# Plot the two test distributions from before
plt.plot(dist_pts, trunc_norm_pdf(dist_pts, mu_1, sig_1, 450),
         linewidth=2, color='r', label='1: $\mu$=300,$\sigma$=30')
plt.legend(loc='upper left')

plt.plot(dist_pts, trunc_norm_pdf(dist_pts, mu_2, sig_2, 450),
         linewidth=2, color='g', label='2: $\mu$=400,$\sigma$=70')
plt.legend(loc='upper left')

# Plot the MLE estimated distribution
plt.plot(dist_pts, trunc_norm_pdf(dist_pts, mu_MLE, sig_MLE, 450),
         linewidth=2, color='k', label='3: $\mu$=558,$\sigma$=176')
plt.legend(loc='upper left')
plt.savefig('MLEplots.png')
```

La línea negra es la distribución óptima dado los datos que se tienen. Pero véase cómo es que la curva verde pareciera ajustar mejor. Lo que pienso que se ha hecho es ajustar una distribución gaussiana unos datos que no son gaussianos y es por eso que parece ser que la curva negra no es la óptima.

```python
print('Log-likelihood 1: ', log_lik_norm(pts, mu_1, sig_1, 450))
print('Log-likelihood 2: ', log_lik_norm(pts, mu_2, sig_2, 450))
print('MLE log-likelihood 3: ', log_lik_norm(pts, mu_MLE, sig_MLE, 450))
```

Así que MLE si es el valor más grande.

Ahora veamos si el $likelihood$ se comporta bien observando una cuadrícula sobre los posibles valores de $\mu$ y $\sigma$ para los datos dados.

```python
cmap1 = matplotlib.cm.get_cmap('winter')

mu_vals = np.linspace(530, 591, 50)
sig_vals = np.linspace(150, 200, 50)
lnlik_vals = np.zeros((50, 50))
for mu_ind in range(50):
    for sig_ind in range(50):
        lnlik_vals[mu_ind, sig_ind] = log_lik_norm(pts, mu_vals[mu_ind],
                                                   sig_vals[sig_ind], 450)

mu_mesh, sig_mesh = np.meshgrid(mu_vals, sig_vals)

fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection='3d')
ax.plot_surface(sig_mesh, mu_mesh, lnlik_vals, rstride=8,cstride=1, cmap=cmap1)
#fig.colorbar(surf, shrink=0.5, aspect=10)
ax.scatter(176.70611057,558.361993,-911.1245098593138,color='k', marker='o',linewidth='5',cmap=cmap1)
ax.set_title('Log likelihood for values of mu and sigma',fontsize=20)
ax.set_xlabel(r'$\sigma$',fontsize=20)
ax.set_ylabel(r'$\mu$',fontsize=20)
ax.set_zlabel(r'log likelihood',fontsize=20)
```

Ahora se debe analisar la varianza de los valores estimados. Como lo hace el autor es a través del método de Empirical Hessian estimator, es decir,
\begin{equation}
VAR(\hat{\theta}) = - H^{-1}(\hat{\theta})
\end{equation}

```python
results_uncstr
vcv_mle = results_uncstr.hess_inv
stderr_mu_mle = np.sqrt(vcv_mle[0,0])
stderr_sig_mle = np.sqrt(vcv_mle[1,1])
print('VCV(MLE) = ', vcv_mle)
print('Standard error for mu estimate = ', stderr_mu_mle)
print('Standard error for sigma estimate = ', stderr_sig_mle)
```

Pero uno debe obtener la varianza del estimador a través de Cramer-Rao.
\begin{equation}
VAR(\hat{\theta}) \geq \dfrac{1}{n} H^{-1}(\hat{\theta})
\end{equation}

```python
results_uncstr
vcv_mle = results_uncstr.hess_inv
stderr_mu_mle = np.sqrt(vcv_mle[0,0])/np.sqrt(161)
stderr_sig_mle = np.sqrt(vcv_mle[1,1])/np.sqrt(161)
print('VCV(MLE) = ', vcv_mle)
print('Standard error for mu estimate = ', stderr_mu_mle)
print('Standard error for sigma estimate = ', stderr_sig_mle)
```

Si se toma en cuenta sólo el hessiano, entonces no se debe despreciar la curva verde, ya que esta está dentro del error de los parámetros.

```python
lb_mu_95pctci = mu_MLE - 2 * stderr_mu_mle
print('mu_2=', mu_2, ', lower bound 95% conf. int.=', lb_mu_95pctci)

lb_sig_95pctci = sig_MLE - 2 * stderr_sig_mle
print('sig_2=', sig_2, ', lower bound 95% conf. int.=', lb_sig_95pctci)
```

```python
log_lik_h0 = log_lik_norm(pts, mu_1, sig_1, 450)
log_lik_mle = log_lik_norm(pts, mu_MLE, sig_MLE, 450)
LR_val = 2 * (log_lik_mle - log_lik_h0)
pval_h0 = 1.0 - sts.chi2.cdf(LR_val, 2)
print('chi squared of H0 with 2 degrees of freedom p-value = ', pval_h0)
```
