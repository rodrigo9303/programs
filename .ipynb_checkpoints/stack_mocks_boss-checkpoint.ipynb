```python
import numpy as np
import matplotlib.pyplot as plt
import fitsio 
from glob import glob
%matplotlib inline
```

```python
fdir = '/home/rodrigo/Documentos/maestria/ultimo_año/datos_boss_dr9/BOSSLyaDR9_cat.fits'
```

```python
DM = fitsio.FITS(fdir)
#DM
```

```python
fm = fitsio.read(fdir,'BOSSLyaDR9_cat')
fm.dtype.descr
```

No all spectra have a the fitted PCA continuum very well. If $CONT$_$FLAG = 2$ the continuum is badly fit. Then, I'm going to check How many spectra have a bad fit.

```python
flag2 = []
for i in range(len(fm['CONT_FLAG'])):
    if fm['CONT_FLAG'][i] == 2:
        flag2.append(i)
```

```python
print('The number of spectra with badle fit = ',len(flag2))
```

But the paper said that: should yield rms continuum  errors  well  under∼$10\%.$, and as after I will make the mean flux, I shouldn't discard this these flags.

```python
print('Number of QSO spectra = ',len(fm['Z_VI']))
```

```python
print('redshift min. = {} \nredshift max = {}'.format(np.min(fm['Z_VI']),np.max(fm['Z_VI'])))
```

I must match redshifts of the BOSSLyaDR9_cat.dat file with each speclya file. For this, I will try to build the path where the file is located.

```python
plates = []
mjds = [] # MJD (Modified Julian date of observation)
fibers = []
zss = []
prueba = []
for i in range(len(fm['Z_VI'])):
    if 2.2 <= fm['Z_VI'][i] < 2.4 :
        plates.append(fm['PLATE'][i])
        mjds.append(fm['MJD'][i])
        fibers.append(fm['FIBERID'][i])
        zss.append(fm['Z_VI'][i])
        prueba.append('BOSSLyaDR9_spectra/{}/speclya-{}-{}-{}.fits'.format(fm['PLATE'][i],fm['PLATE'][i],fm['MJD'][i],str(fm['FIBERID'][i]).zfill(4)))
```

```python
print('long. plates = {} \nlong. mjds = {} \nlong. fibers = {} \nnumber of redshifts = {}'.format(len(plates),len(mjds),len(fibers),len(zss)))
```

Now, I read the first 2000 spctra for this bin in z.

```python
wave_rf = []
continuos = []
flujos = []
for i in range(len(zss))[:2000]:
    specfile = '/home/rodrigo/Documentos/maestria/ultimo_año/datos_boss_dr9/{}'.format(prueba[i])
    sp = fitsio.read(specfile)
    wave = []
    continuo = []
    flujo = []
    for j in range(len(sp['FLUX'])):
        wave.append(10**(sp['LOGLAM'][j]))
        continuo.append(sp['CONT'][j])
        flujo.append(sp['FLUX'][j])
    wave_rf.append(wave/(1+zss[i]))
    continuos.append(continuo)
    flujos.append(flujo)
```

$\textbf{Here, I begin make the normalization.}$

```python
lmin=1216.0
lmax=1546.0

integral = []
continuo_nor = []
for i in range(len(wave_rf)):
    sum1=0
    sum2=0
    flag=0
    for j in range(len(wave_rf[i])):
        if lmin <= wave_rf[i][j] < lmax :
            sum1+=(continuos[i][j])*(wave_rf[i][j+1]-wave_rf[i][j])
        elif (wave_rf[i][j] > lmax):
            break
    integral.append(sum1)
    continuo_nor.append(continuos[i]/sum1)
```

```python
print('The total integrals = {}'.format(len(integral)))
```

$\textbf{Now, I will get the mean flux of the specta.}$

For that, I need to get the minimum wavelength for the continuum flux.

```python
mini = []
for i in range(len(integral)):
    mini.append(np.min(wave_rf[i]))
print('The minimum = ',np.min(mini))
```

```python
count = np.zeros(1600-int(np.min(mini))+1)
mean= np.zeros(1600-int(np.min(mini))+1)
sumflux = np.zeros(1600-int(np.min(mini))+1)
for k in range(len(wave_rf)):
    for l in range(len(wave_rf[k])):
        if int(np.min(mini)) < wave_rf[k][l] <= 1600:
            co = int(wave_rf[k][l])-int(np.min(mini))
            count[co]+=1
            sumflux[co]+=continuo_nor[k][l]
            mean = sumflux/count
        if wave_rf[k][l] > 1600:
            break
```

```python
xmean = np.linspace(int(np.min(mini)),1600,1600-int(np.min(mini))+1) 
#len(xmean)
```

```python
plt.figure(figsize=(15,9))
plt.title('BOSS spectra QSO',fontsize = 18)
for k in range(len(wave_rf))[:50]:
    plt.plot(wave_rf[k],continuo_nor[k], color='b',linewidth=2.0,alpha=0.5)
plt.plot(xmean,mean, color='r',label='Average',linewidth=2.0,alpha=1)
plt.axvline(1216, color = 'k',label=('Ly-$\\alpha$ Peak'),linewidth=2.5,alpha=1)
plt.axvline(1546, color = 'y',label=('CIV Peak'),linewidth=2.5,alpha=1)
#plt.axvline(1049,linewidth=2.5)
plt.axhline(0.0022, color = 'g',label=('0.0022'),linewidth=2.5,alpha=1)
plt.xlim(1030,1600)
#plt.ylim(-5,12)
plt.xlabel('Wavelength$_{R.F.}$  ($\AA$)',fontsize = 18)
plt.ylabel('Normalized Flux      ($10^{-17} ergs/s/cm^{2}/\AA$)',fontsize = 18)
plt.legend(fontsize='x-large')
plt.grid()
#plt.savefig('stacking_boss.png')
```

$\textbf{Now, I will get the mean flux of the specta.}$

Now, I will get the mean flux of the specta.
For that, I need to get the minimum wavelength for the continuum flux.

```python
mini2 = []
for i in range(len(integral2)):
    mini2.append(np.min(wave_rf2[i]))
print('The minimum = ',np.min(mini2))
```

```python
count2 = np.zeros(1600-int(np.min(mini2))+1)
mean2= np.zeros(1600-int(np.min(mini2))+1)
sumflux2 = np.zeros(1600-int(np.min(mini2))+1)
vari = np.zeros(1600-int(np.min(mini2))+1)
varian = np.zeros(1600-int(np.min(mini2))+1)
for k in range(len(wave_rf2)):
    for l in range(len(wave_rf2[k])):
        if int(np.min(mini2)) < wave_rf2[k][l] <= 1601:
            co2 = int(wave_rf2[k][l])-int(np.min(mini2))
            count2[co2]+=1
            sumflux2[co2]+=continuo_nor2[k][l]
            #mean2 = sumflux2/count2
            mean2 = np.array(sumflux2/count2)
            vari[co2]+=(continuo_nor2[k][l]-mean2[co2])**2
            varian = np.sqrt(vari/count2)
        if wave_rf2[k][l] > 1600:
            break
```

```python
wave_rf4 = []
continuos4 = []
flujos4 = []
for i in range(len(zss4)):
    specfile4 = '/home/rodrigo/Documentos/maestria/ultimo_año/datos_boss_dr9/{}'.format(prueba4[i])
    sp4 = fitsio.read(specfile4)
    wave4 = []
    continuo4 = []
    flujo4 = []
    for j in range(len(sp4['FLUX'])):
        wave4.append(10**(sp4['LOGLAM'][j])) #I change LOGWAVE to wavelength.
        continuo4.append(sp4['CONT'][j])
        flujo4.append(sp4['FLUX'][j])
    wave_rf4.append(wave4/(1+zss4[i])) #I change the wavelength to wavelength in the rest-frame.
    continuos4.append(continuo4)
    flujos4.append(flujo4)
```

$\textbf{Normalizing}$

```python
lmin=1216.0
lmax=1546.0

integral4 = []
continuo_nor4 = []
for i in range(len(wave_rf4)):
    sum1=0
    sum2=0
    flag=0
    for j in range(len(wave_rf4[i])):
        if lmin <= wave_rf4[i][j] < lmax :
            sum1+=(continuos4[i][j])*(wave_rf4[i][j+1]-wave_rf4[i][j])
        elif (wave_rf4[i][j] > lmax):
            break
    integral4.append(sum1)
    continuo_nor4.append(continuos4[i]/sum1)
```

```python
print('The total integrals = {}'.format(len(integral4)))
```

Getting the mean-flux

```python
mini4 = []
for i in range(len(integral4)):
    mini4.append(np.min(wave_rf4[i]))
print('The minimum = ',np.min(mini4))
```

```python
count4 = np.zeros(1600-int(np.min(mini4))+1)
mean4 = np.zeros(1600-int(np.min(mini4))+1)
sumflux4 = np.zeros(1600-int(np.min(mini4))+1)
for k in range(len(wave_rf4)):
    for l in range(len(wave_rf4[k])):
        if int(np.min(mini4)) < wave_rf4[k][l] <= 1601:
            co4 = int(wave_rf4[k][l])-int(np.min(mini4))
            count4[co4]+=1
            sumflux4[co4]+=continuo_nor4[k][l]
            mean4 = sumflux4/count4
        if wave_rf4[k][l] > 1600:
            break
```
